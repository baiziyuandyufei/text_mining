```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(6, 8))
plt.subplot(2,1,1)
f = 0.5
t = np.linspace(-10,10,10000)
y_t = np.sinc(2*f*t)
plt.plot(t,y_t, label='sinc(pi*t) f=%.1fHz' % f)
plt.axhline(y=0,ls=":") #
plt.axhline(y=np.max(y_t),ls=":") #
plt.axhline(y=np.min(y_t),ls=":") #
plt.axvline(x=0,ls=":") #
plt.xlim(-10,10)
plt.ylim(-0.4,1.2)
plt.title("sinc(pi*t) f=%.1fHz" % f)

plt.subplot(2,1,2)
f=10
y_t = np.sinc(2*f*t)
plt.plot(t,y_t,label='sinc(2*pi*f*t)')
plt.axhline(y=0,ls=":") #
plt.axhline(y=np.max(y_t),ls=":") #
plt.axhline(y=np.min(y_t),ls=":") #
plt.axvline(x=0,ls=":") #
plt.xlim(-10,10)
plt.ylim(-0.4,1.2)
plt.title("sinc(2*pi*f*t) f=%dHz" % f)
plt.show()

plt.figure(figsize=(13, 6))
plt.subplot(2,2,1)
fs=0.5*f
Ts=1/fs
n = np.arange(-10,10,Ts)
y_n = np.sinc(2*f*n)
plt.scatter(n, y_n)
plt.xlim(-10,10)
plt.ylim(-0.4,1.2)
plt.title("sinc(2*pi*f*t) sample, fs=%dHz" % fs)

plt.subplot(2,2,2)
fs=2*f
Ts=1/fs
n = np.arange(-10,10,Ts)
y_n = np.sinc(2*f*n)
plt.scatter(n, y_n)
plt.xlim(-10,10)
plt.ylim(-0.4,1.2)
plt.title("sinc(2*pi*f*t) sample, fs=%dHz" % fs)

plt.subplot(2,2,3)
fs=5*f
Ts=1/fs
n = np.arange(-10,10,Ts)
y_n = np.sinc(2*f*n)
plt.scatter(n, y_n)
plt.xlim(-10,10)
plt.ylim(-0.4,1.2)
plt.title("sinc(2*pi*f*t) sample, fs=%dHz" % fs)

plt.subplot(2,2,4)
fs=20*f
Ts=1/fs
n = np.arange(-10,10,Ts)
y_n = np.sinc(2*f*n)
plt.scatter(n, y_n)
plt.xlim(-10,10)
plt.ylim(-0.4,1.2)
plt.title("sinc(2*pi*f*t) sample, fs=%dHz" % fs)

plt.show()
```
