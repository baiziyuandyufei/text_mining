```python
import numpy as np
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
plt.figure(figsize=(13, 6))
alpha1 = 0.1
alpha2 = 0.2
x = np.linspace(0,10,100)
f_1 = alpha1*np.exp(-alpha1*x)
f_2 = alpha2*np.exp(-alpha2*x)
plt.subplot(2,2,1)
plt.plot(x, f_1, label='f1(x)')
plt.plot(x, f_2, label='f2(x)')
plt.legend()
plt.title("alph2>alpha1>0")

alpha1 = 0.2
alpha2 = 0.1
x = np.linspace(0,10,100)
f_1 = alpha1*np.exp(-alpha1*x)
f_2 = alpha2*np.exp(-alpha2*x)
plt.subplot(2,2,2)
plt.plot(x, f_1, label='f1(x)')
plt.plot(x, f_2, label='f2(x)')
plt.legend()
plt.title("alpha1>alpha2>0")

alpha1 = -0.1
alpha2 = -0.2
x = np.linspace(0,10,100)
f_1 = alpha1*np.exp(-alpha1*x)
f_2 = alpha2*np.exp(-alpha2*x)
plt.subplot(2,2,3)
plt.plot(x, f_1, label='f1(x)')
plt.plot(x, f_2, label='f2(x)')
plt.legend(loc = 'best')
plt.title("alpha2<alpha1<0")

alpha1 = -0.2
alpha2 = -0.1
x = np.linspace(0,10,100)
f_1 = alpha1*np.exp(-alpha1*x)
f_2 = alpha2*np.exp(-alpha2*x)
plt.subplot(2,2,4)
plt.plot(x, f_1, label='f1(x)')
plt.plot(x, f_2, label='f2(x)')
plt.legend(loc = 'best')
plt.title("alpha1<alpha2<0")
plt.show()
```
