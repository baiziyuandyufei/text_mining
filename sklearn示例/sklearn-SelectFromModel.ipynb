```python
# 训练集文本
texts = ['Chinese Beijing Chinese',
         'Chinese Chinese Shanghai',
         'Chinese Macao',
         'Tokyo Japan Chinese']
# 训练集各文本类别,-1为China类，1为Japan类
y = [-1, -1, -1, 1]
# 文本TF法向量化表示
from sklearn.feature_extraction.text import CountVectorizer
tf = CountVectorizer()
X = tf.fit_transform(texts)
print(tf.get_feature_names())
print(X.toarray())
```

```python
# 训练朴素贝叶斯模型，并输出模型系数
from sklearn.naive_bayes import MultinomialNB
m = MultinomialNB(fit_prior=True)
m.fit(X, y)
print(tf.get_feature_names())
print(m.coef_)
print(m.feature_log_prob_)
```

```python
# 模拟基于模型的特征选择
# 计算均值并将其作为阈值
import numpy as np
threshold = np.mean(m.coef_)
print(threshold)
# 输出选择出的特征
for term, w_value in zip(tf.get_feature_names(), m.coef_[0]):
    if w_value >= threshold:
        print(term, end=' ')
```

```python
# 使用sklearn.SelectFromModel进行特征选择
from sklearn.feature_selection import SelectFromModel
help(SelectFromModel)
```

```python
# 实例化基于模型特征选择实例
fs = SelectFromModel(estimator=MultinomialNB())
# 训练分类模型并执行特征选择
fs.fit(X, y)
# 获取特征选择阈值
print(fs.threshold_)
# 输出选择出的特征，显然是错误的
print([w[1] for w in zip(fs.get_support(), tf.get_feature_names()) if w[0]])
```
