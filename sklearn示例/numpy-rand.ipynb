```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 正选序列幅度
A = np.sqrt(2)
# 角频率
w = np.pi/16
# 时间序列
n = np.arange(100)
# 均匀分布噪声均值
mean = 0
# 均匀分布噪声序列功率（方差）
p = 0.1
# 均匀分布u(mean, p)噪声序列
u_n = (np.random.rand(50000)[25000:25000+len(n)] + (mean-0.5))*np.sqrt(12*p)
# 加入噪声的正弦序列
x_n = A*np.sin(w*n) + u_n
plt.plot(n, x_n)
plt.show()
```
