```python
# 生成训练数据
import numpy as np
X = np.random.randint(5, size=(6, 100))
y = np.array([1, 2, 3, 4, 5, 6])
```

```python
# 训练数据包含6个样本，每个样本100维
X
```

```python
# 训练数据的类别标记
y
```

```python
# 实例化多项式朴素贝叶斯模型
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.get_params()
```

```python
# 训练多项式朴素贝叶斯模型
clf.fit(X, y)
```

```python
# 预测训练集中的第3个样本的类别
clf.predict(X[2:3])
```

```python
# 每个类别的先验概率值
clf.intercept_
```

```python
# 打印每个类别的多项式参数，也就是词语在每个个类别下的频率值
# 其中每行表示一个类别，每列表示一个词语，因此wij，表示词语j在类别i文档集中出现的频率
clf.coef_
```
