# 2. legend()方法

```python
from numpy.random import rand
# 防止图例中的中文显示不出来设置字体实例
from matplotlib.font_manager import FontProperties
# 不同操作系统字体的文件会不同
font = FontProperties(fname='C:\Windows\Fonts\simkai.ttf')
%matplotlib inline

# 生成一个子图图对象，ax每画一次都会对应一种图例
fig, ax = plt.subplots()
for color, color_name in zip(['red', 'green', 'blue'], ['红色', '绿色', '蓝色']):
    n = 750
    x, y = rand(2, n)
    scale = 200.0 * rand(n)
    ax.scatter(x, y, c=color, s=scale, label=color_name,
               alpha=0.3, edgecolors='none')
# 设置设置字体参数
ax.legend(prop=font)
ax.grid(True)

plt.show()
```
